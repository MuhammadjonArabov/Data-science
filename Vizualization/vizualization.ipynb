```python
cols = ["petal-length", "petal-width", "sepal-length", "sepal-width"]
plt.hist([data["petal-length"], data["petal-width"], data["sepal-length"], data["sepal-width"]], label=[cols[0], cols[1], cols[2], cols[3]])
plt.legend()
```

```python
plt.hist(data["petal-length"])
plt.hist(data["petal-width"])
plt.hist(data["sepal-length"])
plt.hist(data["sepal-width"])
plt.show()
```

```python
plt.hist(data["petal-length"], alpha=0.2)
plt.hist(data["petal-width"], alpha=0.3)
plt.hist(data["sepal-length"], alpha=0.6)
plt.hist(data["sepal-width"], alpha=0.9)
plt.show()
```

```python
fig, ax = plt.subplots(2, 2, figsize=(20, 10))
ax[0,0].hist(data["petal-length"])
ax[0,1].hist(data["petal-width"])
ax[1,0].hist(data["sepal-length"])
ax[1,1].hist(data["sepal-width"])
plt.show()
```

```python
fig, ax = plt.subplots(2, 2, figsize=(20, 10))
ax[0,0].hist(data["petal-length"], color='lime')
ax[0,1].plot(data["petal-width"])
ax[1,0].plot(data["sepal-length"], c="m")
ax[1,1].boxplot(data["sepal-width"])
plt.show()
```

```python
plt.scatter(
    data["petal-length"],
    data["petal-width"]
)
```

```python
colors = {
    "Iris-setosa": "red",
    "Iris-versicolor": "blue",
    "Iris-virginica": "black",  
}

plt.scatter(
    data["petal-length"],
    data["petal-width"],
    c=data["class"].map(colors)
)
plt.xlabel(cols[0])
plt.ylabel(cols[1])
plt.title("Iris Vizualization")
plt.show()
```

```python
fig, ax = plt.subplots(2, 2, figsize=(15, 10))
A = [data["sepal-length"][data["class"] == 'Iris-setosa'], data["sepal-length"][data["class"] == 'Iris-virginica'], data["sepal-length"][data["class"] == 'Iris-versicolor']]
B = [data["sepal-width"][data["class"] == 'Iris-setosa'], data["sepal-width"][data["class"] == 'Iris-virginica'], data["sepal-width"][data["class"] == 'Iris-versicolor']]
C = [data["petal-length"][data["class"] == 'Iris-setosa'], data["petal-length"][data["class"] == 'Iris-virginica'], data["petal-length"][data["class"] == 'Iris-versicolor']]
D = [data["petal-width"][data["class"] == 'Iris-setosa'], data["petal-width"][data["class"] == 'Iris-virginica'], data["petal-width"][data["class"] == 'Iris-versicolor']]

ax[0, 0].boxplot(A, widths = 0.7)
ax[0, 0].set_title(cols[0])
ax[0, 1].boxplot(B, widths = 0.7)
ax[0, 1].set_title(cols[1])
ax[1, 0].boxplot(C, widths = 0.7)
ax[1, 0].set_title(cols[2])
ax[1, 1].boxplot(D, widths = 0.7)
ax[1, 1].set_title(cols[3])
```
