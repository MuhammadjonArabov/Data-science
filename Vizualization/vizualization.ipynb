```python
plt.hist(data["sepal-length"])
```

```python
plt.hist([data["petal-length"], data["petal-width"], data["sepal-length"], data["sepal-width"]])
```

```python
cols = ["petal-length", "petal-width", "sepal-length", "sepal-width"]
plt.hist([data["petal-length"], data["petal-width"], data["sepal-length"], data["sepal-width"]], label=[cols[0], cols[1], cols[2], cols[3]])
plt.legend()
```

```python
plt.hist(data["petal-length"])
plt.hist(data["petal-width"])
plt.hist(data["sepal-length"])
plt.hist(data["sepal-width"])
plt.show()
```

```python
plt.hist(data["petal-length"], alpha=0.2)
plt.hist(data["petal-width"], alpha=0.3)
plt.hist(data["sepal-length"], alpha=0.6)
plt.hist(data["sepal-width"], alpha=0.9)
plt.show()
```

```python
fig, ax = plt.subplots(2, 2, figsize=(20, 10))
ax[0,0].hist(data["petal-length"])
ax[0,1].hist(data["petal-width"])
ax[1,0].hist(data["sepal-length"])
ax[1,1].hist(data["sepal-width"])
plt.show()
```

```python
fig, ax = plt.subplots(2, 2, figsize=(20, 10))
ax[0,0].hist(data["petal-length"], color='lime')
ax[0,1].plot(data["petal-width"])
ax[1,0].plot(data["sepal-length"], c="m")
ax[1,1].boxplot(data["sepal-width"])
plt.show()
```
