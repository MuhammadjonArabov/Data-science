```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
plt.plot([0, 4, 1, 5, 3])

plt.show()
```

```python
plt.plot([3, 2, 5, 9], [4, 1, 7, 9])

plt.title("Data")

plt.xlabel("data x")

plt.ylabel("data y")

plt.show()
```

```python
plt.figure(figsize=(2, 4))

plt.plot([1, 2, 3], [2, 5, 6], 'ro')

plt.axis([0, 5, 0, 8])

plt.annotate("data o'sishi", (1, 7))

plt.show()
```

```python
plt.hist([1, 4, 2, 0])

plt.show()
```

```python
plt.clf()

x = np.arange(5)

y = [8.4, 5.2, 4, 7, 9]

plt.xticks(x, ("Ali", "Vali", "Jamol", "Sobir", "Holid"))

plt.bar(x,y,color=["lime", "r", "k", "m", "b"])

plt.show()
```

```python
data = {
    "Red 0": np.random.rand(10),
    "Grn X": np.random.rand(10)
    }

df = pd.DataFrame(data)

df.plot(style=['ro', 'bx'])

plt.show()
```

```python
t = pd.Series(np.random.randn(180), index=pd.date_range(start='1/1/2018', periods=180))

df = pd.DataFrame(np.random.randn(180, 3), index=t.index, columns=['A', 'B', 'C'])

df.cumsum().plot()

plt.show()
```
