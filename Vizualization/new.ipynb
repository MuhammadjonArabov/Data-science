```python
df = pd.read_csv("data.txt")  

# Shtab diagramma 
plt.bar(df["month"], df["record_high"], color="r", label="Record High")
plt.bar(df["month"], df["record_low"], color="b", label="Record Low")

# Chiziqli diagramma 
plt.plot(df["month"], df["avg_high"], color="k", marker="o", label="Avg High")
plt.plot(df["month"], df["avg_low"], color="c", marker="o", label="Avg Low")

plt.legend()

plt.show()

print(df)
```

```python
# 2 ta subplot yaratish, x o'qi bir xil bo'lishi uchun sharex=True
fig, plots = plt.subplots(2, sharex=True)

# Grafik sarlavhasi
fig.suptitle("Sharing x axis")

# Ma'lumotlar
x = range(0, 200, 10)
y = [n**0.8 for n in x]

# Birinchi subplot - chiziqli grafik
plots[0].plot(x, y, color="r")
plots[0].set_ylabel("Line Plot")

# Ikkinchi subplot - scatter (nuqta) grafik
plots[1].scatter(x, y)
plots[1].set_ylabel("Scatter Plot")
plots[1].set_xlabel("X values")

# Grafikni korsatish
plt.show()
```
