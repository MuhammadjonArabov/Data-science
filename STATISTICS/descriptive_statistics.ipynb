Main - o'rta qiymat.  Median — tartiblangan ma’lumotlar to‘plamining o‘rtasidagi qiymat. Moda — eng ko‘p uchragan qiymat.                                                                                                                        Variance (Dispersiya – tarqoqlik o‘lchovi)
Har bir qiymatning o‘rtacha qiymatdan qanchalik uzoqligini ko‘rsatadi. Katta dispersiya → katta tarqoqlik.
Formula:    Variance= ∑(xi −Mean)**2/n

```python
import numpy as np
from scipy import stats

data = [70, 75, 80, 85, 90, 95, 100]

print("Mean: ", np.mean(data))
print("Median: ", np.median(data))

mode_result = stats.mode(data, keepdims=True)
print("Mode:", mode_result.mode[0])
print("Variance: ", np.var(data))
```

Normal Distribution (Normal Taqsimot)

```python
import matplotlib.pyplot as plt

mu = 70  # o'rtacha ball
sigma = 10 # standarat og'ish
data_normal = np.random.normal(mu, sigma, 1000)

plt.hist(data_normal, bins=30, edgecolor='black', alpha=0.7)
plt.title('Normat Distribution (Mu=70, Sigma=10)')
plt.xlabel('Ball')
plt.ylabel('Frekans')
plt.show()
```

Poisson Distribution (Poisson Taqsimot)

```python
lambda_val = 4
data_poisson = np.random.poisson(lambda_val)

plt.hist(data_poisson, bins=20, edgecolor='blue', alpha=0.7)
plt.title(f'Poisson Distribution (Lambda={lambda_val})')
plt.xlabel('Avtobuslar soni')
plt.ylabel('Frekans')
plt.show()
```

Binomial Distribution (Binomial Taqsimot)

```python
n = 10
p = 0.5
data_binomail =np.random.binomial(n, p, 1000)

plt.hist(data_binomail, bins=range(0, 12), edgecolor='yellow', alpha=0.7)
plt.title(f'Binomial Distribution (n={n}, p={p})')
plt.xlabel('Yuzlar soni')
plt.ylabel('Frekans')
plt.show
```
