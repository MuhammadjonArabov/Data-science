```python
import numpy as np

x = np.array([6, 7, 8, 2])
y = np.array([1, 2, 3, 4])

# qo'shish
print(f"add = {np.add(x, y)}")

# ko'paytirish
print(f"subtract = {np.subtract(x, y)}")

# bo'lish
print(f"divide = {np.divide(x, y)}")

# butun qismini bo'lish
print(f"floor_divide = {np.floor_divide(x, y)}")

# qoldiqni olish
print(f"mode = {np.mod(x, y)}")

# darajaga oshirish
print(f"power = {np.power(x, 2)}")

# qoldiqni olish
print(f"remainder = {np.remainder(x,y)}")

# absalut qiymat
z = np.array([1, -2, -4, -5, 8])
print(f"abs = {np.abs(z)}")

# manfiy qiymatga aylantiradi
print(f"negative = {np.negative(x)}")

# exsponensiyal funksiya (e^x)
print(f"exp = {np.exp(y)}")

# natural logarifm (ln)
print(f"log = {np.log(x)}")

# logarifm
print(f"log10 = {np.log10(y)}")

# logarifm 2 asosli
print(f"log2 = {np.log2(x)}")

# Sinus
print(f"sin = {np.sin(x)}")

# Cosinus
print(f"Cos = {np.cos(y)}")

# Tang
print(f"Tang = {np.tanh(x)}")

# ArcSinus
print(f"arcsinh = {np.arcsinh(x)}")

e = np.array([0.5, -0.5])
print(f" arctanh = {np.arctanh(e)}")  

# Radianni gradusga o'girish
print(f"degrees = {np.degrees(np.pi)}")  

# Gradusni radianlarga o'girish
print(f" radians = {np.radians(180)}") 

# Gipotenuzani topish
print(f"hypot = {np.hypot(3, 4)}")  
```
